#  splitting dataset in x and y

```python
x = v.drop(columns = ["Chance of Admit "])
y = v["Chance of Admit "]
```

#   train test split

```python
from sklearn.model_selection  import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y,test_size = 0.2 , random_state = 2)
```

 # scaling by using minmax scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# ANN Architecture

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import backend as K
import numpy as np
import random

# -----------------------------
# Seed set for reproducibility
# -----------------------------
seed = 42
tf.random.set_seed(seed)
np.random.seed(seed)
random.seed(seed)

# Custom RÂ² metric
def r2_score(y_true, y_pred):
    ss_total = tf.reduce_sum(tf.square(y_true - tf.reduce_mean(y_true)))
    ss_residual = tf.reduce_sum(tf.square(y_true - y_pred))
    return 1 - ss_residual / (ss_total + K.epsilon())

# Custom RMSE metric
def rmse(y_true, y_pred):
    return tf.sqrt(tf.reduce_mean(tf.square(y_true - y_pred)))

# Build simple model
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=7))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))

# Compile the model
model.compile(loss='mse',
              optimizer='adam',
              metrics=['mae', rmse, r2_score])

# Early stopping callback
early_stop = EarlyStopping(monitor='val_loss', restore_best_weights=True)

# Train the model
history = model.fit(
    x_train, y_train,
    epochs=120,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# XGBoost regression model tuning
xgb_reg = XGBRegressor(
    n_estimators= 400,
    learning_rate=0.01,  # regression me 0.1 best hota usually
    max_depth= 3,
    random_state=42,
    objective='reg:squarederror'  # regression ke liye
)

# Fitting model to train dataset
xgb_reg.fit(x_train, y_train)

# Prediction on test dataset
y_pred15 = xgb_reg.predict(x_test)

# Metrics for regression

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

mse = mean_squared_error(y_test, y_pred15)
mae = mean_absolute_error(y_test, y_pred15)
r2 = r2_score(y_test, y_pred15)
rmse = np.sqrt(mse)  # RMSE calculate

print("MSE:", mse)
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
